Taking all inputs on one line, separated by spaces (or other delimiters)
If you want the user to type all their values on a single line, you can use input() once and then use the .split() method to separate the values into individual variables. By default, .split() will split a string by whitespace.

Problem: [C. Simple Calculator](https://codeforces.com/group/MWSDmqGsZm/contest/219158/problem/C)

Solution:

````python
n = input()
a,b = n.split()
a = int(a) # type casting
b = int(b)

print(f"{a} + {b} = {a+b}")
print(f"{a} * {b} = {a*b}")
print(f"{a} - {b} = {a-b}")
````


 the input() function always reads what you type as a string. Even if you type numbers like '5' and '10', Python treats them as characters, not numerical values. When you then use .split(), a becomes the string '5' and b becomes the string '10'.

To fix this, you need to explicitly convert these strings to numbers using int() (for integers) or float() (for decimal numbers) before you can perform any calculations.
