```python
print("Hello, future Python master!")

# Note
# Colon : The parentheses are followed by a colon, which signifies the start of the function's code block.
# Indentation: The lines of code that make up the function's body must be indented. This indentation is crucial in Python, as it defines the scope of the function.

# --- 1. Variables --- #
# Variables are like containers for storing data. Let's create a few.
my_name = "Alice"  # This is a string variable
my_age = 30        # This is an integer variable
my_height = 1.75   # This is a float variable
is_student = True  # This is a boolean variable

print(f"\n1. Variables:") #The f you see before the opening quote in f"\n1. Variables:" stands for f-string. It's a powerful and concise way to embed expressions inside string literals in Python. It's short for "formatted string literal."#
print(f"Name: {my_name}, Age: {my_age}, Height: {my_height}m, Student: {is_student}")

# --- 2. Conditionals (if/elif/else) --- #
# Conditionals allow your code to make decisions based on certain conditions.
print(f"\n2. Conditionals:")
if my_age < 18:
    print(f"{my_name} is a minor.")
    # print(f"{my_name} is a minor.")
elif my_age >= 18 and my_age < 65:
    print(f"{my_name} is an adult.")
    print(f"{my_name} is an adult.") #python supports mutliple lines with same indent#
else:
    print(f"{my_name} is a senior citizen.")

# Let's check if the person is a student AND an adult
if is_student and my_age >= 18:
    print(f"{my_name} is an adult student.")

# --- 3. Loops (for loop and while loop) --- #
# Loops help you repeat actions. We'll use a 'for' loop and a 'while' loop.

print(f"\n3. Loops:")
# For loop: Iterating through a sequence (like numbers in a range or characters in a string)
print("For loop (counting to 3):")
for i in range(1, 4): # range(1, 4) generates numbers 1, 2, 3
    print(f"Count: {i}")

print("\nWhile loop (counting down from 3):")
count_down = 3
while count_down > 0:
    print(f"T-minus {count_down}...")
    count_down -= 1 # This is the same as count_down = count_down - 1
print("Lift off!")

# --- 4. User-Defined Functions --- #
# Functions are blocks of organized, reusable code that perform a single, related action.
# They help make your code modular and easier to manage.

print(f"\n4. User-Defined Functions:")

#def: This stands for "define." It's the keyword that tells Python you are about to create a new function.

def greet_person(name): # This function takes one argument: 'name'
    """This function prints a personalized greeting."""
    print(f"Hello there, {name}! Welcome to Python programming.")

def add_numbers(num1, num2): # This function takes two arguments and returns their sum
    """This function takes two numbers and returns their sum."""
    return num1 + num2

# Calling our functions:
greet_person(my_name)
greet_person("Bob")

result_sum = add_numbers(10, 5)
print(f"The sum of 10 and 5 is: {result_sum}")

# You can also use variables in function calls
number_a = 7
number_b = 3
final_sum = add_numbers(number_a, number_b)
print(f"The sum of {number_a} and {number_b} is: {final_sum}")

# --- 5. User Input --- #
# The input() function allows you to get data from the user.
print(f"\n5. User Input:")
user_name = input("Please enter your name: ")
greet_person(user_name)

print("\nKeep practicing! That's the best way to learn.")
```
