Problem: [G. Summation from 1 to N](https://codeforces.com/group/MWSDmqGsZm/contest/219158/problem/G)

Solution: 

```python
# Problem G: Summation from 1 to N
# Given a number N. Print the summation of the numbers that is between 1 and N (inclusive).

def solve():
    try:
        n = int(input())
        # Use formula n*(n+1)/2 to avoid TLE for large N
        print(n * (n + 1) // 2)
    except (EOFError, ValueError):
        pass

if __name__ == '__main__':
    solve()
```
