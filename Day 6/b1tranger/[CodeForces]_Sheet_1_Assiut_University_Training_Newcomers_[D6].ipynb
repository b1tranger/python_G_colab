```python
def binSearch(arr,size,target):
    low = 0
    high = size-1
    while low <=high:
     mid = low + (high-low)//2 # Changed / to // for integer division
     if(arr[mid]==target):
      return 1
     elif(arr[mid]<target):
      low = mid+1
     else:
      high = mid-1
    return 0

n = input() # size, test case
a = n.split()
size = int(a[0])
run = int(a[1])
str_arr = input()
arr = str_arr.split() # input regardless of "size" value
for j in range(run):
    target = input()
    target = int(target)
    flag = 0
    flag = binSearch(arr,size,target)
    if (flag==1):
      print("found")
    else:
      print("not found")
```

```python
# Define a function that takes a list as an argument
def process_list(my_list):
  print("Inside the function:")
  print(f"Original list: {my_list}")

  # You can iterate through the list
  print("Elements:")
  for item in my_list:
    print(item)

  # You can modify the list (if mutable)
  # For example, adding an element
  my_list.append("new_item")
  print(f"List after modification: {my_list}")
  print("-" * 20)

# Declare some lists (arrays)
my_numbers = [1, 2, 3, 4, 5]
my_strings = ["apple", "banana", "cherry"]

# Call the function, passing the lists
print("Calling process_list with my_numbers:")
process_list(my_numbers)

print("Calling process_list with my_strings:")
process_list(my_strings)

# Observe that my_numbers was modified by the function (lists are mutable)
print(f"my_numbers after function calls: {my_numbers}")
```

### Explanation:

1.  **Function Definition**: The function `process_list(my_list)` is defined, specifying `my_list` as its parameter. When you call this function, whatever list you pass to it will be referred to as `my_list` inside the function.
2.  **Passing by Reference (Implicitly)**: In Python, when you pass a mutable object like a list to a function, you are essentially passing a reference to that object. This means that if you modify the list *inside* the function (e.g., using `append()`, `remove()`, etc.), those changes will be reflected in the original list outside the function. This is demonstrated by the `my_numbers` list being modified after `process_list` is called.

You can achieve something similar to C++'s fixed-size array input in Python, though Python lists are dynamically sized. Here are two common ways to handle looped input to create a list of a predetermined or implied size:

### Method 1: Reading the size `N`, then `N` individual elements

This method is closer to how you might pre-declare an array size in C++ and then fill it element by element in a loop. In Python, you'd typically initialize an empty list or a list of a specific size with placeholder values, and then populate it.

```python
# Read the desired size of the list
list_size = int(input("Enter the number of elements: "))

# Initialize an empty list
my_list = []

# Loop to take individual elements as input
print(f"Enter {list_size} elements, one by one:")
for i in range(list_size):
    element = input(f"Element {i+1}: ")
    my_list.append(element)

print(f"Your list (string elements): {my_list}")

# If you need integer elements, you can convert them during input:
# my_int_list = []
# for i in range(list_size):
#     element = int(input(f"Element {i+1} (integer): "))
#     my_int_list.append(element)
# print(f"Your list (integer elements): {my_int_list}")
```

### Method 2: Reading all elements on a single line

This is often more convenient in competitive programming. You read a single line of space-separated values and then use the `split()` method to create a list of strings. You can then convert these strings to integers if needed.

```python
# Read a single line of space-separated numbers
input_line = input("Enter elements separated by spaces: ")

# Split the line into a list of strings
string_list = input_line.split()

print(f"Your list (string elements): {string_list}")

# To convert to a list of integers:
# int_list = [int(x) for x in string_list]
# print(f"Your list (integer elements): {int_list}")

# Note: The cell 'ol3NhUdGPgBX' in your notebook already uses this method:
# n = input()
# a = n.split()
# size = len(a)
```
