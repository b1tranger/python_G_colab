Problem: [D. Difference](https://codeforces.com/group/MWSDmqGsZm/contest/219158/problem/D)

Solution:

```python
n = input()
a,b,c,d = n.split()
a = int(a)
b = int(b)
c = int(c)
d = int(d)
x = (a*b)-(c*d)
print(f"Difference = {x}")
```

Problem: [E. Area of a Circle](https://codeforces.com/group/MWSDmqGsZm/contest/219158/problem/E)

Solution:

```python
pi = 3.141592653
R = input()
R = float(R)
A = pi*R*R
print(A)
# print(f"{A:.9f}")
```

Problem: [F. Digits Summation](https://codeforces.com/group/MWSDmqGsZm/contest/219158/problem/F)

Solution:

To get the digit at a certain position, you can also convert the number to a string and use string indexing. Remember that string indexing is zero-based (the first character is at index 0).


```python
n = input() # input 2 values
a,b = n.split()
# keep them as string
# now we need to find the last digits of a,b

length_a = len(a) # returns the number of indecies of a
length_b = len(b)

# print(length_a)
# print(length_b)

last_digit_a = a[length_a-1] # returns the last digit of a
last_digit_b = b[length_b-1]

# print(last_digit_a)
# print(last_digit_b)

# a = last_digit_a
# b = last_digit_b

# a = int(a)
# b = int(b)

a = int(last_digit_a)
b = int(last_digit_b)

print(a+b)

```
